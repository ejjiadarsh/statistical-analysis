**Performing descriptive statistics**

Descriptive statistics are used to summarize and organize data in a way that makes it easy to understand. They provide a simple snapshot of the characteristics of a dataset. Descriptive statistics are divided into three main types:

1. Measures of Central Tendency:
These statistics describe the center or typical value of a dataset.

Mean: The average of all data points.
Median: The middle value when data points are ordered.
Mode: The most frequent value in a dataset.
2. Measures of Dispersion:
These show the spread or variability in the data.

Range: The difference between the highest and lowest values.
Variance: The average of the squared differences from the mean.
Standard Deviation: The square root of the variance, which shows the average amount of variability in the dataset.
Interquartile Range (IQR): The range within which the central 50% of data points lie (difference between the 75th and 25th percentile).
3. Measures of Shape:
These describe the distribution of the data.

Skewness: Indicates whether the data is symmetrically distributed or skewed to one side (left or right).
Kurtosis: Describes the "tailedness" of the distribution—whether it has heavier or lighter tails compared to a normal distribution.
Descriptive statistics are often visualized using charts like histograms, bar charts, and box plots to offer better insight into the dataset.

```python
#calculate the basic descriptive statistics
print("MEAN:\n\n",df.mean())
print("\nMEDIAN:\n\n",df.median())
print("\nMODE:\n\n",df.mode().iloc[0])
print("\nSTANDARD DEVIATION:\n\n",df.std())
print("\nVARIANCE:\n\n",df.var())


#additional descriptive statistics
print("\nRANGE:\n\n",df.max()-df.min())
print("\nSKEWNESS:\n\n",df.skew())
print("\nKURTOSIS:\n\n",df.kurt())
```

**Performing Inferential Statistics**

Inferential statistics is a branch of statistics that involves making predictions or inferences about a population based on a sample of data. While descriptive statistics summarize and describe the characteristics of a dataset, inferential statistics go a step further by using the data to make generalizations and test hypotheses about a larger population.

```python
from scipy import stats#
#example data:uterine_contractions
fetal_health_values=df['fetal_health']
#population mean
population_mean = 1.50
#perform one sample test
t_stat, p_value = stats.ttest_1samp(fetal_health_values, population_mean)

print(f"T-static:{t_stat}")
print(f"p-vales:{p_value}")
```

**Confidence Intervals**

Confidence intervals (CIs) are a statistical tool used to estimate the range within which a population parameter (such as a mean or proportion) is likely to lie, given a sample of data. The interval is calculated from the data and is associated with a confidence level, typically expressed as a percentage (e.g., 95%).


```python
import numpy as np
from scipy import stats
#Sample mean and standard error for fetal_health
sample_mean=np.mean(fetal_health_values)
standard_error=stats.sem(fetal_health_values)
# Compute 95% confidence interval for fetal_health
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)
print(f"95% Confidence Interval for fetal_health: {confidence_interval}")
```

** Regression Analysis**

Regression analysis is a statistical technique used to examine the relationship between one or more independent variables (predictors) and a dependent variable (outcome). It helps to understand how the dependent variable changes when any one of the independent variables is varied, while keeping the others fixed.


```python
import statsmodels.api as sm
# Define independent variable (add constant for intercept)
X = sm.add_constant(df['fetal_health'])
#define dependent variable
y = df['uterine_contractions']
#fit linear regression model
model = sm.OLS(y, X).fit()
#print model summary
print(model.summary())
```
